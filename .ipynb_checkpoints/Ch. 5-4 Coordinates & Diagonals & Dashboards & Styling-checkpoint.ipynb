```python
import pandas as pd
movies = pd.read_csv('D:\\Users\\Minerva\\Programming\\Udemy\\Python A-Z Python For Data Science With Real Exercises\\6. Advanced Visualization\\MovieRatings.csv')
movies.columns = ['Film','Genre','RottenRating','AudienceRating','Budget','Year']
movies.head()
from matplotlib import pyplot as plt
import seaborn as sbn
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Facet Grid Scatter Plot
g = sbn.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=.5, edgecolor='black')
g = g.map(plt.scatter, 'RottenRating', 'AudienceRating', **kws)
```

```python
# Lets add diagonals for easier visualisation and change the x and y range
g = sbn.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=.5, edgecolor='black')
g = g.map(plt.scatter, 'RottenRating', 'AudienceRating', **kws)
g.set(xlim=(0,100),ylim=(0,100))
# g.axes accesses each facet .flat treats it as a single array for simplicity
for axes in g.axes.flat:
    # axes.plot((xCoord),(yCoord), args)
    axes.plot((0,100),(0,100), c='gray', ls='--')
# legend
g.add_legend()
```

```python
sbn.set_style('darkgrid')
f, axes = plt.subplots(2,2, figsize=(15,15))
k1 = sbn.kdeplot(movies.Budget, movies.AudienceRating, ax=axes[0,0])
k2 = sbn.kdeplot(movies.Budget, movies.RottenRating, ax=axes[0,1])
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
v1 = sbn.violinplot(data=movies[movies.Genre=='Drama'], x='Year',y='RottenRating', ax=axes[1,0])
k1 = sbn.kdeplot(movies.RottenRating, movies.AudienceRating, shade=True, shade_lowest=False, cmap='Blues', ax=axes[1,1])
k1 = sbn.kdeplot(movies.RottenRating, movies.AudienceRating, cmap='Blues', ax=axes[1,1])
plt.show()
```

```python
# -------------------------------------------------------------------------------------------------------------
# Adding titles & stuff for finishing touches
```

```python
movies.Genre = movies.Genre.astype('category')
```

```python
list1 = list()
labels = list()
for genre in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==genre].Budget)
    labels.append(genre)
    
sbn.set_style('whitegrid')
fig, ax = plt.subplots()
fig.set_size_inches(11.7,8.27)
h = plt.hist(list1, bins=30, stacked=True, rwidth=1, label=labels)
plt.title('Movie Budget Distribution', fontsize=25, color='DarkBlue', fontname='Console')
plt.ylabel('Number of Movies', fontsize=25, color='Red')
plt.xlabel('Budget', fontsize=25, color='Green')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(frameon=True, shadow=True, prop={'size':20})
plt.show()
```
