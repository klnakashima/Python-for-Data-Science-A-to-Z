```python
sbn.set(style='darkgrid',palette='muted',color_codes=True)
plt.figure(figsize=(15,9))
ax =sbn.boxplot(data=mov2, x='Genre', y='USPercent', orient='v',color='lightgray', )
plt.setp(ax.artists, alpha=0.5)

sbn.stripplot(x='Genre', y='USPercent', data=mov2, jitter=True, size=10, linewidth=0, hue='Studio', alpha=0.7)
ax.axes.set_title('Domestic Gross % by Genre', fontsize=20)
ax.set_xlabel('Genre', fontsize=20)
ax.set_ylabel('US Gross %', fontsize=20)
ax.legend(bbox_to_anchor=(1.05,1), loc=2)
```
