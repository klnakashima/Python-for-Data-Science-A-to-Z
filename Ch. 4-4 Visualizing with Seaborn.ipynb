```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline
plt.rcParams['figure.figsize'] = 16,9

import warnings
warnings.filterwarnings('ignore')

stats = pd.read_csv('D:\\Users\\Minerva\\Programming\\Udemy\\Python A-Z Python For Data Science With Real Exercises\\5. Data Frames\\P4-Demographic-Data.csv')
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers','IncomeGroup']
stats.columns
```

```python
stats.head()
```

```python
# Distribution
vis1 = sbn.distplot(stats['InternetUsers'])
```

```python
vis1 = sbn.distplot(stats['InternetUsers'], bins=40)
```

```python
# Boxplots
vis2 = sbn.boxplot(data=stats, x='IncomeGroup',y='BirthRate')
```

```python
# Seaborn Gallery Online - Good resource
```

```python
# linear model
vis3 = sbn.lmplot(data=stats, x='InternetUsers',y='BirthRate', fit_reg=False)
```

```python
vis4 = sbn.lmplot(data=stats, x='InternetUsers',y='BirthRate', fit_reg=False, \
                 hue='IncomeGroup', size=10, aspect=1.5)
```

```python
# --------------------------------------------------------------------------------------------------------------------
# Keyword Arguments in Python
```

```python
# How to change marker size
# scatter_kws uses ply.scatter with extra arguments you can use
# From matplotlib documentation we see scatter_kws has argument s that requires a dict item like this:
vis5 = sbn.lmplot(data=stats, x='InternetUsers',y='BirthRate', fit_reg=False, \
                 hue='IncomeGroup', size=8, aspect=1.2, scatter_kws={"s":100})
```
