```python
gList = list()
gLabels = list()
for genre in movies.Genre.cat.categories:
    gList.append(movies[movies.Genre == genre].Budget)
    gLabels.append(genre)
    
plt.hist(gList, bins=30, stacked=True, rwidth=1, label=gLabels)
plt.legend()
plt.show()
```
