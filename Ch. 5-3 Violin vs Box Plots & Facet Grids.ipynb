```python
import pandas as pd
movies = pd.read_csv('D:\\Users\\Minerva\\Programming\\Udemy\\Python A-Z Python For Data Science With Real Exercises\\6. Advanced Visualization\\MovieRatings.csv')
movies.columns = ['Film','Genre','RottenRating','AudienceRating','Budget','Year']
movies.head()
from matplotlib import pyplot as plt
import seaborn as sbn
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#w = sbn.violinplot(data=movies, x='Genre', y='RottenRating')
w1 = sbn.violinplot(data=movies, x='Genre', y='RottenRating')
```

```python
w2 = sbn.boxplot(data=movies, x='Genre', y='RottenRating')
```

```python
w3 = sbn.violinplot(data=movies[movies.Genre=='Drama'], x='Year', y='RottenRating')
```

```python
w4 = sbn.boxplot(data=movies[movies.Genre=='Drama'], x='Year', y='RottenRating')
```

```python
# -------------------------------------------------------------------------------------------------------
# Creating a Facet Grid
```

```python
vis1 = sbn.lmplot(data=movies, x='RottenRating', y='AudienceRating', fit_reg=False, hue='Genre', size=7, aspect=1.2)
```

```python
# This is too complicated to read
# Facet grid different than sublots, where you subplots populate grids
# FG you split up one visualization with rules that you specify
g = sbn.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
```

```python
g = sbn.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
g = g.map(plt.scatter, 'RottenRating', 'AudienceRating')
```

```python
# Can split up other types of graphs
g = sbn.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
g = g.map(plt.hist, 'Budget')
```

```python
# Back to scatter, let's add a kwargs
g = sbn.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=.5, edgecolor='black')
g = g.map(plt.scatter, 'RottenRating', 'AudienceRating', **kws)
```
